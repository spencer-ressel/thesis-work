```python
import os
import numpy as np
from glob import glob
import xarray as xr
```

```python
os.chdir('/home/disk/eos7/sressel/research/thesis-work/python/numerical_solver/output/Ahmed-21/')
print(f"{os.getcwd()}")

all_directories = os.listdir('/home/disk/eos7/sressel/research/thesis-work/python/numerical_solver/output/Ahmed-21/')


for directory in all_directories:
    if not os.path.isdir(directory):
        all_directories.remove(directory)
    # Rossby_downsampled_files = glob(f"./*downsampled*")

for directory in all_directories:     
    print(f"Directory: {directory}")
    print(f"======================")
    for k in range(1,6):
        for c in range(1,12):
            files = sorted(glob(
                f"{directory}/Rossby-wave_initial-condition/*k={k}.0*downsampled*of{c}.nc"
            ))
            
            if not files:
                # print(f"k={k}, c={c}: No matching files")
                pass
            else:
                print(f"k={k}")
                data = []
                for file in files:
                    print(file)
                    data.append(xr.load_dataset(file))
                compiled_file = xr.concat(data, dim='time')
                # print(compiled_file)
                compiled_file.to_netcdf(
                    f"{directory}/Rossby-wave_initial-condition/k={k}.0_m=1_Rossby-wave_downsampled-model-data_compiled.nc")
                print(f"\n")

print("done")
```
