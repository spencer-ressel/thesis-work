# Vertical Mass Flux example

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.arange(0, 10e3, 1)
rho0 = 1.225
rho = rho0*np.exp(-z/5e3)

# w = z*np.exp(-z**2)
w = np.sin(np.pi*z/np.max(z))
M = w*rho
M_prime = np.gradient(M, z/1e3)

plt.figure()
plt.style.use('bmh')
plt.plot(rho, z)
plt.plot(w, z)
plt.plot(M, z, label='Mass Flux')
plt.plot(M_prime, z, color='k')
# plt.plot(1/M[1000:-1000], z[1000:-1000])
# plt.plot(1/M, z)
plt.yticks(ticks=z[::1000], labels=np.arange(0, 10, 1))
plt.ylabel('km')
plt.legend()

plt.axvline(x=0, color='gray', ls=':', lw=2)

plt.show()
```
